Customized Series

```python
data = [10, 20, 30, 40, 50, 60]
series1 = pd.Series(data)
series1
```

```python
data = [10, 20, 30, 40, 50, 60]
series2 = pd.Series(data, index=['a', 'b', 'c', 'd', 'e', 'f'])
series2
```

```python
s1 = pd.Series([10, 20, 30], index = ['a', 'b', 'c'])
s2 = pd.Series([5, 15, 25], index = ['b', 'c', 'd'])

result = s1 + s2

print(result)
```

```python
a1 = np.array([10, 20, 30])
a2 = np.array([5, 15, 25])

print(a1 + a2)
```

iloc vs loc -> to access the values in a given dataframe

```python
data.iloc[[1,2,5,9,12],[3,4,6,8,9]]
```

iloc -> you can access any row/column using the index numbers only

```python
first_name_column = data.iloc[:, 1]
```

```python
first_name_column
```

Access the column as 'street_name'

```python
street_name_column = data.iloc[:, 6]
```

```python
street_name_column
```

Access the rows having index number [1, 2, 3] and the column name [firstName, lastName, Gender]

```python
data.iloc[1:4, [1, 2, 4]]
```

loc -> to access the elements using the label-based indexing

```python
data.loc[1:4, ['first_name', 'last_name', 'gender']]
```
